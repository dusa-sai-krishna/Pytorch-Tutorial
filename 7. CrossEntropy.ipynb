- Used to calculate the loss in the cases of multi- classification


- For Pytorch, keep these things in mind


```python
import torch
import torch.nn as nn
```

```python
loss=nn.CrossEntropyLoss()

# Let's consider a situation where three classes are present. Here we are considering only one sample
y = torch.tensor([0]) # correct class labyel

# 1 sample, raw values for all possible classes 
y_pred_good = torch.tensor([[2,1,0.1]],dtype=torch.float32)
y_pred_bad = torch.tensor([[0.5,2,0.3]],dtype=torch.float32)


l1 = loss(y_pred_good,y)
l2 = loss(y_pred_bad,y)

print(l1,l2)
```

```python
# If we are giving data of more than one sample , let's say three samples
loss=nn.CrossEntropyLoss()

# Let's consider a situation where three classes are present. Here we are considering only one sample
y = torch.tensor([0,1,2]) # correct class labyel

# 1 sample, raw values for all possible classes 
y_pred_good = torch.tensor([[2,1,0.1],[2,1,0],[1,3,1]],dtype=torch.float32)
y_pred_bad = torch.tensor([[0.5,2,0.3],[2,1,0.1],[1,3,0.1]],dtype=torch.float32)


l1 = loss(y_pred_good,y)
l2 = loss(y_pred_bad,y)

print(l1,l2)
```
